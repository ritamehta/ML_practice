```python
# converting finalscore into categorical variable - # Excellent - 15 to 20 Good - 10 to 14 Fair - 0 to 9
student['finalgrade'] = "na"
student.loc[(student.finalscore >= 15) & (student.finalscore <= 20), 'finalgrade'] = 'excellent' 
student.loc[(student.finalscore >= 10) & (student.finalscore <= 14), 'finalgrade'] = 'good' 
student.loc[(student.finalscore >= 0) & (student.finalscore <= 9), 'finalgrade'] = 'fair' 
student.head(5)
```
